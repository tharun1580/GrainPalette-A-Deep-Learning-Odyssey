```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
data_reg = pd.DataFrame({
    'X1': np.random.randint(1, 100, 100),
    'X2': np.random.randint(1, 100, 100),
    'Y': np.random.randint(1, 200, 100)
})
X_reg = data_reg[['X1', 'X2']]
y_reg = data_reg['Y']
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Larger dataset for classification
data_clf = pd.DataFrame({
    'X1': np.random.randint(0, 50, 200),
    'X2': np.random.randint(0, 50, 200),
    'Y': np.random.randint(0, 2, 200)
})
X_clf = data_clf[['X1', 'X2']]
y_clf = data_clf['Y']
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)

# Function to plot actual vs predicted using line graph
def plot_results(y_test, y_pred, title):
    plt.plot(range(len(y_test)), y_test, color='blue', label='Actual', marker='o')
    plt.plot(range(len(y_pred)), y_pred, color='red', label='Predicted', marker='x')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
def simple_linear_regression():
    model = LinearRegression()
    model.fit(X_train_reg[['X1']], y_train_reg)
    y_pred = model.predict(X_test_reg[['X1']])
    print("Simple Linear Regression MSE:", mean_squared_error(y_test_reg, y_pred))
    plot_results(y_test_reg, y_pred, "Simple Linear Regression")
simple_linear_regression()
```

```python
def multiple_linear_regression():
    model = LinearRegression()
    model.fit(X_train_reg, y_train_reg)
    y_pred = model.predict(X_test_reg)
    print("Multiple Linear Regression MSE:", mean_squared_error(y_test_reg, y_pred))
    plot_results(y_test_reg, y_pred, "Multiple Linear Regression")
multiple_linear_regression()
```

```python
def decision_tree_regressor():
    model = DecisionTreeRegressor()
    model.fit(X_train_reg, y_train_reg)
    y_pred = model.predict(X_test_reg)
    print("Decision Tree Regressor MSE:", mean_squared_error(y_test_reg, y_pred))
    plot_results(y_test_reg, y_pred, "Decision Tree Regressor")
decision_tree_regressor()
```

```python
def random_forest_regressor():
    model = RandomForestRegressor()
    model.fit(X_train_reg, y_train_reg)
    y_pred = model.predict(X_test_reg)
    print("Random Forest Regressor MSE:", mean_squared_error(y_test_reg, y_pred))
    plot_results(y_test_reg, y_pred, "Random Forest Regressor")
random_forest_regressor()
```

```python
def logistic_regression():
    model = LogisticRegression()
    model.fit(X_train_clf, y_train_clf)
    y_pred = model.predict(X_test_clf)
    print("Logistic Regression Accuracy:", accuracy_score(y_test_clf, y_pred))
logistic_regression()
```

```python
def decision_tree_classifier():
    model = DecisionTreeClassifier()
    model.fit(X_train_clf, y_train_clf)
    y_pred = model.predict(X_test_clf)
    print("Decision Tree Classifier Accuracy:", accuracy_score(y_test_clf, y_pred))
decision_tree_classifier()
```

```python
def naive_bayes():
    model = GaussianNB()
    model.fit(X_train_clf, y_train_clf)
    y_pred = model.predict(X_test_clf)
    print("Naive Bayes Accuracy:", accuracy_score(y_test_clf, y_pred))
naive_bayes()
```

```python
def support_vector_classifier():
    model = SVC()
    model.fit(X_train_clf, y_train_clf)
    y_pred = model.predict(X_test_clf)
    print("SVC Accuracy:", accuracy_score(y_test_clf, y_pred))
support_vector_classifier()
```

```python
def xgboost_classifier():
    model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
    model.fit(X_train_clf, y_train_clf)
    y_pred = model.predict(X_test_clf)
    print("XGBoost Classifier Accuracy:", accuracy_score(y_test_clf, y_pred))
xgboost_classifier()
```
