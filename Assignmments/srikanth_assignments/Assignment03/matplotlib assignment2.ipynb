```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


plt.figure(figsize=(18, 20))
np.random.seed(42)

#line chart(sales trend)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [35, 42, 38, 45, 52, 48]
plt.subplot(5, 2, 1)
plt.plot(months, sales, marker='o', color='royalblue')
plt.title('Monthly Sales Trend 2023')
plt.xlabel('Months')
plt.ylabel('Sales (in $1000)')
```

```python
# 2. Bar Chart (Product Comparison)
products = ['Laptops', 'Tablets', 'Phones', 'Accessories']
units_sold = [45, 32, 68, 56]
plt.subplot(5, 2, 2)
plt.bar(products, units_sold, color='lightgreen')
plt.title('Product Sales Performance')
plt.xticks(rotation=45)
```

```python
#3. Histogram (Age Distribution)
ages = np.random.normal(35, 10, 1000)
plt.subplot(5, 2, 3)
plt.hist(ages, bins=15, edgecolor='black', color='salmon')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
```

```python
# 4. Scatter Plot (Height vs Weight)
height = np.random.normal(170, 10, 100)
weight = height * 0.4 + np.random.normal(60, 8, 100)
plt.subplot(5, 2, 4)
plt.scatter(height, weight, alpha=0.6, color='purple')
plt.title('Height vs Weight Correlation')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

```

```python
# 5. Pie Chart (Market Share)
companies = ['Company A', 'Company B', 'Company C', 'Others']
market_share = [35, 30, 25, 10]
plt.subplot(5, 2, 5)
plt.pie(market_share, labels=companies, autopct='%1.1f%%', startangle=90)
plt.title('Market Share Distribution')
```

```python
# 6. Box Plot (Exam Scores)
math = np.random.normal(70, 10, 100)
physics = np.random.normal(65, 15, 100)
plt.subplot(5, 2, 6)
plt.boxplot([math, physics], tick_labels=['Math', 'Physics'])
plt.title('Exam Score Distributions')
plt.ylabel('Scores')
plt.show
```

```python
# 7. Heatmap (Correlation Matrix)
data = pd.DataFrame({
    'Sales': sales * 2,
    'Expenses': [20, 22, 25, 23, 28, 24] * 2,
    'Profit': [15, 20, 13, 22, 24, 24] * 2
})
plt.subplot(5, 2, 7)
plt.imshow(data.corr(), cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(3), data.columns)
plt.yticks(range(3), data.columns)
plt.title('Feature Correlation Matrix')
```

```python
# 8. Stacked Bar Chart (Quarterly Sales)
categories = ['Electronics', 'Fashion', 'Home']
Q1 = [20, 35, 30]
Q2 = [25, 32, 28]
Q3 = [30, 28, 35]
Q4 = [35, 25, 32]
plt.subplot(5, 2, 9)
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
bottom = np.zeros(4)
for i, category in enumerate(categories):
    values = [Q1[i], Q2[i], Q3[i], Q4[i]]
    plt.bar(quarters, values, label=category, bottom=bottom)
    bottom += values
plt.title('Quarterly Sales by Category')
plt.legend
```

```python
# 9. Polar Plot (24h Temperature)
hours = np.arange(0, 24)
theta = np.linspace(0, 2*np.pi, 24)
temperature = 10 + 10*np.sin(theta) + np.random.normal(0, 2, 24)
plt.subplot(5, 2, 10, polar=True)
plt.plot(theta, temperature, marker='o', color='darkorange')
plt.title('24 Hour Temperature Variation')
plt.xticks(theta[::2], hours[::2])
plt.grid(True)

plt.tight_layout()
plt.show()


```

```python
# 10. Area Chart (Website Traffic)
days = np.arange(1, 31)
direct = np.random.randint(50, 100, 30)
organic = np.random.randint(100, 200, 30)
paid = np.random.randint(30, 80, 30)
plt.subplot(5, 2, 8)
plt.stackplot(days, direct, organic, paid, 
              labels=['Direct', 'Organic', 'Paid'],
              colors=['skyblue', 'lightgreen', 'pink'])
plt.title('Daily Website Traffic Sources')
plt.xlabel('Days')
plt.ylabel('Visitors')
plt.legend(loc='upper left')
```
