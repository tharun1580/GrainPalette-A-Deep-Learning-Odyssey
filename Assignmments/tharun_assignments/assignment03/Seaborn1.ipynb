```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

sns.set_theme(style="whitegrid")
plt.figure(figsize=(18, 20))

# Sample data generation
np.random.seed(42)
data = pd.DataFrame({
    'Age': np.random.normal(35, 10, 1000),
    'Income': np.random.lognormal(4, 0.4, 1000),
    'Gender': np.random.choice(['Male', 'Female'], 1000),
    'Product': np.random.choice(['A', 'B', 'C'], 1000),
    'Satisfaction': np.random.randint(1, 6, 1000),
    'Temperature': np.random.normal(25, 5, 1000),
    'Species': np.random.choice(['Setosa', 'Versicolor', 'Virginica'], 1000)
})
```

```python
# 1. Distribution Plot (Age Distribution)
plt.subplot(5, 2, 1)
sns.histplot(data=data, x='Age', kde=True, bins=20, color='skyblue')
plt.title('1. Age Distribution Analysis')
```

```python
# 2. Count Plot (Product Popularity)
plt.subplot(5, 2, 2)
sns.countplot(x='Product', data=data, hue ='Product',  palette='viridis' , legend =False)
plt.title('2. Product Popularity Comparison')
```

```python
# 3. Box Plot (Income Distribution by Gender)
plt.subplot(5, 2, 3)
sns.boxplot(x='Gender', y='Income', data=data, hue='Gender', palette='pastel' , legend = False)
plt.title('3. Income Distribution by Gender')
```

```python
# 4. Heatmap (Correlation Matrix)
plt.subplot(5, 2, 4)
corr_matrix = data[['Age', 'Income', 'Satisfaction', 'Temperature']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('4. Feature Correlation Matrix')
```

```python
# 5. Violin Plot (Temperature Distribution by Species)
plt.subplot(5, 2, 5)
sns.violinplot(x='Species', y='Temperature', data=data, hue='Species', inner='quartile', palette='Set2' , legend= False)
plt.title('5. Temperature Distribution by Species')
```

```python
# 6. Pair Plot (Multivariate Relationships)
plt.subplot(5, 2, 6)
sns.pairplot(data=data[['Age', 'Income', 'Satisfaction']], 
            diag_kind='kde', 
            plot_kws={'alpha': 0.6})
plt.title('6. Multivariate Relationships')
plt.gcf().set_size_inches(10, 8)
```

```python
# 7. Line Plot (Time Series with Confidence Interval)
time_data = pd.DataFrame({
    'Month': pd.date_range(start='2023-01', periods=24, freq='ME'),
    'Sales': np.cumsum(np.random.normal(100, 20, 24))
})
plt.subplot(5, 2, 7)
sns.lineplot(x='Month', y='Sales', data=time_data, errorbar=('ci',95), marker='o')
plt.title('7. Monthly Sales Trend with CI')
```

```python
# 8. Swarm Plot (Satisfaction by Product and Gender)
plt.subplot(5, 2, 8)
sns.stripplot(x='Product', y='Satisfaction', hue='Gender', data=data, dodge=True, palette='dark')
plt.title('8. Satisfaction Rating by Product & Gender')
```

```python
# 9. Joint Plot (Age vs Income Relationship)
plt.subplot(5, 2, 9)
sns.jointplot(x='Age', y='Income', data=data, kind='hex', cmap='Blues')
plt.title('9. Age-Income Relationship')
plt.gcf().set_size_inches(10, 8)
```

```python

# 10. Facet Grid (Multi-panel Analysis)
plt.subplot(5, 2, 10)
g = sns.FacetGrid(data, col='Product', hue='Gender', palette='Set1')
g.map(sns.scatterplot, 'Age', 'Income', alpha=0.6)
g.add_legend()
plt.subplots_adjust(top=0.85)
g.fig.suptitle('10. Age-Income Relationship by Product & Gender')

plt.tight_layout()
plt.show()
```
